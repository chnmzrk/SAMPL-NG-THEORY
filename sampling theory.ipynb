```python
from itertools import combinations
import numpy as np
import pandas as pd
stack = [10,16,13,24,21,11,23,18]
def unit(x):
     comb = combinations(stack,x)
     sample = []
     for i in comb:
         sample.append(i)
         samples = {"SAMPLE UNITS" : sample}
         pd.set_option('display.max_rows', None)
     a = pd.DataFrame(samples)
     a.index = np.arange(1,len(a)+1)
     print(a)
     print(f"\nMass Avarage : {np.mean(stack)}\nSample Avarage : {np.mean(sample)}"
     f"\nSample Size : {np.shape(sample)}\n")
     
unit(1)
```
